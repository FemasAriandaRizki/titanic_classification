### Visualisasi Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train["Sex"].value_counts()
```

```python
count_sex = train["Sex"].value_counts()

index_sex = count_sex.index
values_sex = count_sex.values
```

```python
index_sex
```

```python
values_sex
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(index_sex, values_sex, color = "maroon", width = 0.3)

plt.xlabel("jenis kelamin")
plt.ylabel("jumlah penumpang")
plt.title("ChartJumlah Penumpang Berdasarkan Jenis Kelamin")

plt.show()
```

```python
def barchart(column):
  count = train[column].value_counts()

  index = count.index
  values = count.values

  fig = plt.figure(figsize = (10, 5))
  plt.bar(index, values, color = "blue", width = 0.2)
```

```python
barchart("Sex")
```

```python
barchart("Pclass")
```

```python
barchart("SibSp")
```

```python
barchart("Parch")
```

Grouped Barchart

```python
train.head()
```

```python
survived = train.groupby(["Survived", "Sex"]).count()["PassengerId"].reset_index()
```

```python
survived
```

```python
def group_barchart(column):
  survived = train.groupby(["Survived", column]).count()["PassengerId"].reset_index()

  g = sns.catplot(
      data = survived, kind = "bar",
      x = "Survived", y = "PassengerId", hue = column,
      palette = "dark", height = 6, width = 0.2
  )
```

```python
group_barchart("Sex")
```

```python
group_barchart("Pclass")
```

Histogram

```python
fig, ax = plt.subplots(figsize = (10, 7))
ax.hist(train.Age)

plt.show()
```

```python
def hstgrm(column):
  fig, ax = plt.subplots(figsize = (10, 7))
  ax.hist(train[column], color='green', bins=30)

  ax.set_title(f'Histogram of {column}')
  ax.set_xlabel(column)
  ax.set_ylabel('Frequency')

  plt.show()
```

```python
hstgrm("Fare")
```
